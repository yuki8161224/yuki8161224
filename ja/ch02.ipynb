## 10. 行数のカウント
ファイルの行数をカウントせよ。確認にはwcコマンドを用いよ。

```python
count = 0

with open("popular-names.txt") as f:
    for line in f:
        count += 1

print(count)
```

```python
!wc -l popular-names.txt
```

## 11. 先頭からN行を出力
ファイルの先頭N行だけを表示せよ。例えば、N=10として先頭10行を表示せよ。確認にはheadコマンドを用いよ。

```python
N = 10

with open("popular-names.txt") as f:
    for i, line in enumerate(f):
        if i >= N:
            break
        print(line, end="")
```

```python
!head -n 10 popular-names.txt
```
